```python
#Toplulastirma ve Gruplama (Aggregation & Grouping)
#count()--first()--last()--mean()--median()--min()--max()--std()
data = sns.load_dataset("diamonds")
```

```python
data.head()
```

```python
data["price"].mean()
```

```python
data.groupby("color")["price"].mean() #rengine gore fiyatlandirma ortalamasi
```

```python
data.groupby("color").agg({"price":"mean"}) # renklere gore fiyat degiskeninin agg kullanarak ortalamasi, aliskanlik halinde kullanilmasi gereken fonksiyonlar
#Ornegin price degerine birden fazla agg fonksiyonu uygulayacaksak daha kullanisli.
data.groupby("color").agg({"price":["mean", "sum"]}) #fiyatin ortalamasi ve toplamini aldik.
```

```python
data
```

```python
data.groupby("color").agg({"price":["mean", "sum"],
                           "carat":["mean"]})
```

```python
data.groupby("color").agg({"price":["mean"],
                           "carat":["mean"]})
```

```python
dialist=["x","y","z"]
data.groupby(dialist).agg({"price":["mean","sum"],
                           "carat":["mean"]})
```

```python
dialist=["x","y","z","depth"]
data.groupby(dialist).agg({"price":["mean","sum"],
                           "carat":["mean"]}).tail()
```

```python
data.groupby(dialist).agg({
    "price":["mean","sum"],
    "carat":["mean"],
    "clarity": ["count"]}).tail()
```

```python
#PIVOT TABLE: 
# Ozet tablosu -> Pivot table ile geniş veri tabloları kolay bir şekilde özetleyebilir (gruplama), yorumlayabilir, filtreleyebilir, koşullu biçimleme (formatting), sıralayabilir (sort) ve işlemler gerçekleştirebiliriz.
# Verileri kirilimlar acisindan degerlendirmek ve ilgilendigimiz ozet istatistigi bu kirilimlar acisindan gormemizi saglar.
data
```

```python
data.pivot_table("price", "cut", "color") #ilk once kesimlerde hangi degeri gormek istiyoruz (price),indexte yani satirda gormek istedigin (cut), en son olarak sutunda (color)
```

```python
data.pivot_table("price", "cut", "color", aggfunc="std")#degiskenlerin standart sapmasi.
```

```python
data.pivot_table("price", "cut", ["color","clarity"])
```

# **APPLY & LAMBDA**

```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
data = sns.load_dataset("car_crashes")
data.head()
```

```python
data["total1"] = data["total"]*2
data["total2"] = data["total"]*5
```

```python
data.head()
```

```python
for col in data.columns:
  if "total" in col:
    data[col] = data[col]/10

data.head()
```

```python
#lambda ve apply kullanarak:
data[["total", "total1", "total2"]].apply(lambda x: x/10).head()
```

```python
#loc kullanarak:
data.loc[:, data.columns.str.contains("total")].apply(lambda x: x/10).head()
```

```python
data.loc[:, data.columns.str.contains("total")].apply(lambda x:(x - x.mean()) / x.std()).head()
```

```python
def std_scaler(col_name):
  return((col_name - col_name.mean()) / col_name.std())
```

```python
#apply bize lamda ile veya hazir fonksiyonlari verilerimize (satir veya sutun) uygulamamizi saglar.
data.loc[:, data.columns.str.contains("total")].apply(std_scaler).head()
data.loc[:,data.columns.str.contains("total")] = data.loc[:, data.columns.str.contains("total")].apply(std_scaler).head()
```

```python
#Merge kullanarak birlestirme islemleri:
import pandas as pd
df = pd.DataFrame({'isim': ['Ethem','Tolgay','Mert','Betül','Busra','Tugba','Kaan','Simal'],
                'yas':[22,23,24,22,22,22,21,21]})
df2 = pd.DataFrame ({'isim': ['Ethem','Tolgay','Mert','Betül','Busra','Tugba','Kaan','Simal'],
        'sehir':['Adiyaman','Ankara','Tekirdag','Artvin','Konya','Kocaeli','Kocaeli','Samsun']})
```

```python
df2
```

```python
pd.merge(df,df2)
```

```python
#hangi degiskene gore birlestirme islemi yapacagini belirtmek istersek :
df3 = pd.merge(df,df2, on = "isim")
```
